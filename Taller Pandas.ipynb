# Transformar el dataset

## df_00

Cree el df_00 con las siguientes características:

Diez columnas (dentro de las diez columnas deben estar la primera y la última columna del dataset original).

Diez filas (las filas deben corresponder a las cinco primeras filas y las cinco últimas filas del dataset original).

Muestre el resultado y las características del DataFrame.

```
# Obtiene las cinco primeras filas del DataFrame original
cinco_primeras_filas = peliculas_disney.head(5)

# Obtiene las cinco últimas filas del DataFrame original
cinco_ultimas_filas = peliculas_disney.tail(5)

# Concatena las dos partes para obtener diez filas en total
df_00 = pd.concat([cinco_primeras_filas, cinco_ultimas_filas])

# Selecciona las diez columnas requeridas (primera y última columna del \
# dataset original más ocho adicionales)
columnas_seleccionadas = (
    peliculas_disney.columns[0:1].tolist() +
    peliculas_disney.columns[-1:].tolist() +
    peliculas_disney.columns[1:9].tolist()
)
df_00 = df_00[columnas_seleccionadas]

# Muestra el resultado
print(df_00)

# Muestra las características del DataFrame df_00
print("\nCaracterísticas del DataFrame df_00:")
print(f"Nombres de las columnas: {df_00.columns}")
print(f"Tipos de datos de las columnas:\n{df_00.dtypes}")
print(f"Índice del DataFrame: {df_00.index}")
print(f"Forma del DataFrame (filas, columnas): {df_00.shape}")
pep8(_ih)
```

## df_05

Calcule el promedio de una de las columnas numéricas de df_02.

Cree el DataFrame df_05 conservando todas las columnas de df_02 y seleccionando los valores superiores al promedio en la columna numérica que tomó como referencia para el ejercicio.

Muestre el resultado y las características del DataFrame.

## df_06

Seleccione una columna de df_02 que contenga una variable categórica (si no hay variables categóricas cree una columna con variables categóricas a partir de una columna numérica de df_02).

Utilice el método **groupby** para analizar la información de df_02 a partir de la variable categórica seleccionada.

Muestre el resultado y las características del DataFrame.

```
# Utiliza groupby para contar registros en cada país
df_06 = df_02.groupby('Sexta').size().reset_index(name='Conteo')

# Muestra el resultado
print(df_06)

# Muestra las características del DataFrame df_06
print("\nCaracterísticas del DataFrame df_06:")
print(f"Nombres de las columnas: {df_06.columns}")
print(f"Índice del DataFrame: {df_06.index}")
print(f"Forma del DataFrame (filas, columnas): {df_06.shape}")
pep8(_ih)
```

## df_07

Cree una tabla dinámica, utilizando el método **pivottable**, para analizar la información de df_02

Muestre el resultado y las características del DataFrame.

```
# Utiliza el método pivot_table para crear una tabla dinámica tomando la \
# columna "Segunda", "Tercera" y "Sexta" y los valores de la 'Octava'
df_07 = df_02.pivot_table(index=["Segunda", "Tercera", "Sexta"],
                          values=['Octava'], aggfunc='mean')

# Muestra el resultado
print(df_07)

# Muestra las características del DataFrame df_07
print("\nCaracterísticas del DataFrame df_07:")
print(f"Nombres de las columnas: {df_07.columns}")
print(f"Índice del DataFrame: {df_07.index}")
print(f"Forma del DataFrame (filas, columnas): {df_07.shape}")
pep8(_ih)
```

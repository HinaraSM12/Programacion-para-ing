## arreglo_02

Cree el arreglo_02 (cinco columnas y diez filas) a partir del dataset original. Todos los datos deben ser números.

Agregue al arreglo_02 lo siguiente:

* una columna con todos los valores iguales a cinco
* una columna con todos los valores iguales a uno
* una columna con números aleatorios provenientes de una distribución normal de media 0 y desviación típica 1
* una columna con números aleatorios enteros en el intervalo 200 a 300
* una columna con números aleatorios binarios (0,1)

Muestre el resultado y las características del arreglo.

```
# Importar la librería random
import random

# Seleccionar columnas numéricas del dataset
id = peliculas_disney[1:11, :1]
running_time = peliculas_disney[1:11, 7:8]
budget = peliculas_disney[1:11, 8:9]
box_office = peliculas_disney[1:11, 9:10]
imdb = peliculas_disney[1:11, 11:12]

# Crear una columna con todos los valores iguales a cinco
columna_cinco = np.full((10, 1), 5)

# Crear una columna con todos los valores iguales a uno
columna_uno = np.full((10, 1), 1)

# Crear una columna con números aleatorios de una distribución\
# normal (media 0, desviación estándar 1)
columna_normal = np.random.normal(0, 1, size=(10, 1))

# Crear una columna con números aleatorios enteros en el intervalo\
# [200, 300]
columna_enteros = np.random.randint(200, 301, size=(10, 1))

# Crear una columna con números aleatorios binarios (0 o 1)
columna_binarios = np.random.randint(0, 2, size=(10, 1))

# Concatenar todas las columnas horizontalmente para formar el arreglo_02
arreglo_02 = np.hstack((id, running_time, budget, box_office, imdb,
                        columna_cinco, columna_uno, columna_normal,
                        columna_enteros, columna_binarios))

# Muestra el resultado y las características del arreglo_02
print(f"arreglo_02: \n {arreglo_02}")
print("\nCaracterísticas del arreglo arreglo_02:")
print(f"La forma del arreglo arreglo_02 es: {arreglo_02.shape}")
print(f"La dimensión del arreglo arreglo_02 es: {arreglo_02.ndim}")
print(f"Número de filas: {arreglo_02.shape[0]}")
print(f"Número de columnas: {arreglo_02.shape[1]}")
pep8(_ih)
```

# Operaciones con dataset

## arreglo_03

Cree el arreglo_03 como la matriz transpuesta del arreglo_02 obtenido en el ejercicio anterior.

Encuentre la inversa del arreglo_03 y multiplíquela por el arreglo_03.

Muestre los arreglos obtenidos y las características de cada uno de ellos.

```
# Calcula la matriz transpuesta (arreglo_03)
arreglo_03 = arreglo_02.T

# Convierte todos los elementos del arreglo a tipo 'float64'
arreglo_03 = arreglo_03.astype(np.float64)

# Encuentra la inversa de arreglo_03
inversa_arreglo_03 = np.linalg.inv(arreglo_03)

# Multiplica la inversa por arreglo_03
resultado_multiplicacion = np.dot(inversa_arreglo_03, arreglo_03)

# Muestra el arreglo 3 y sus características
print(f"arreglo_03 (Transpuesta de arreglo_02): \n {arreglo_03}")
print("\nCaracterísticas de arreglo_03:")
print(f"La forma del arreglo_03 es: {arreglo_03.shape}")
print(f"La dimensión del arreglo_03 es: {arreglo_03.ndim}")
print(f"Número de filas: {arreglo_03.shape[0]}")
print(f"Número de columnas: {arreglo_03.shape[1]} \n")

# Muestra la inversa del arreglo 3 y sus características
print(f"Inversa del arreglo_03: \n {inversa_arreglo_03}")
print("\nCaracterísticas de inversa_arreglo_03:")
print(f"La forma del arreglo inversa_arreglo_03 es: \
{inversa_arreglo_03.shape}")
print(f"La dimensión del arreglo inversa_arreglo_03 es: \
{inversa_arreglo_03.ndim}")
print(f"Número de filas: {inversa_arreglo_03.shape[0]}")
print(f"Número de columnas: {inversa_arreglo_03.shape[1]} \n")

# Muestra la multiplicación inversa_arreglo_03 x arreglo_03 del arreglo 3\
# y sus características
print(f"Resultado de la multiplicación inversa_arreglo_03 x arreglo_03:\
\n {resultado_multiplicacion}")
print("\nCaracterísticas del resultado de la multiplicación:")
print(f"La forma del arreglo resultado_multiplicacion es: \
{resultado_multiplicacion.shape}")
print(f"La dimensión del arreglo resultado_multiplicacion es: \
{resultado_multiplicacion.ndim}")
print(f"Número de filas: {resultado_multiplicacion.shape[0]}")
print(f"Número de columnas: {resultado_multiplicacion.shape[1]}")
pep8(_ih)
```

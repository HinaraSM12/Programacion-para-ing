## gráfico 00

## gráfico 02

Seleccione cuatro columnas numéricas del dataset y haga un gráfico tipo **stackplot** (apilable). Coloque título al gráfico, etiquete los dos ejes e incluya una leyenda dentro del gráfico.

```
# Columnas seleccionadas
indice = peliculas_disney['Índice']
duracion = peliculas_disney['Running time (int)']
puntaje = peliculas_disney['metascore']
internet_movie_data_base = peliculas_disney['imdb']

# Crear el gráfico de áreas apiladas
plt.stackplot(peliculas_disney.index, indice, duracion, puntaje,
              internet_movie_data_base, labels=['Índice', 'Duración',
                                                'Puntaje', 'IMDB'])

# Etiquetas de los ejes y título
plt.xlabel('Índice')
plt.ylabel('Valores de índice, duración, punntaje e imdb')
plt.title('Gráfico de áreas apiladas entre \n Índice, duración, puntaje \
y el imdb de cada película ')
plt.legend(loc='upper left')

# Mostrar el gráfico
plt.show()

# Comprobar que esté en formato pep8
pep8(_ih)
```

## gráfico 03

Cree una figura que incluya los gráficos construidos en los puntos anteriores en un mismo espacio.

```
# Seleccionar las columnas numéricas del dataset necesarias
index = peliculas_disney['Índice']
presupuesto = peliculas_disney['Budget (float)']
duracion = peliculas_disney['Running time (int)']
puntaje = peliculas_disney['metascore']
internet_movie_data_base = peliculas_disney['imdb']

# Crear un diccionario con los datos de cada columna para luego usarlo\
# en el gráfico de scatter
datos_scatter = {'index': index,
                 'presupuesto': presupuesto,
                 'duracion': duracion,
                 'puntaje': puntaje,
                 'internet_movie_data_base': internet_movie_data_base}

# Crear una figura y una cuadrícula de subgráficos (2 filas y 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Gráfico plot
axs[0, 0].plot(index, presupuesto, marker='o',  linestyle='-',
               color='blue', alpha=0.7)
axs[0, 0].set_title('Gráfico de línea entre \n el índice cada película y\
                    su presupuesto')

# Gráfico scatter
axs[0, 1].scatter('index', 'presupuesto', 'duracion', 'puntaje',
                  data=datos_scatter)
axs[0, 1].set_title('Gráfico Scatter entre el índice, \n presupuesto,\
                    duración y puntaje de la película')
axs[0, 1].legend()

# Gráfico de áreas apiladas
axs[1, 0].stackplot(peliculas_disney.index, index, duracion, puntaje,
                    internet_movie_data_base,
                    labels=['Índice', 'Duración', 'Puntaje', 'IMDB'])
axs[1, 0].set_title('Gráfico de áreas apiladas entre\n Índice, duración,\
puntaje y el imdb de cada película')
axs[1, 0].legend()

# Eliminar el espacio vacío en el cuarto gráfico
axs[1, 1].axis('off')

# Ajustar el diseño de los subgráficos
plt.tight_layout()

# Mostrar la figura
plt.show()

# Comprobar que esté en formato pep8
pep8(_ih)
```

## gráfico 07

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de torta (**pie chart**). Coloque título al gráfico y etiquete los segmentos del gráfico.

```
# Seleccionar las columnas numéricas del dataset necesarias
duracion = peliculas_disney['Running time (int)']
puntaje = peliculas_disney['metascore']
internet_movie_data_base = peliculas_disney['imdb']

# Suma de los valores en cada columna
suma_columna1 = duracion.sum()
suma_columna2 = puntaje.sum()
suma_columna3 = internet_movie_data_base.sum()

# Etiquetas para las porciones del gráfico
etiquetas = ['Duración', 'Puntaje', 'IMDB']

# Valores para cada porción del gráfico de torta
valores = [suma_columna1, suma_columna2, suma_columna3]

# Colores para cada porción del gráfico
colores = ['blue', 'orange', 'green']

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(valores, labels=etiquetas, colors=colores, autopct='%1.1f%%',
        startangle=140)
plt.title('Gráfico de torta entre duración, puntaje e imdb de las \
películas')
plt.show()

# Comprobar que esté en formato pep8
pep8(_ih)
```

## gráfico 08

Seleccione las columnas que considere pertinentes del dataset y haga un mapa de calor (**heatmap**). Coloque título al gráfico y etiquete los dos ejes.

```
# Importar librerías
import seaborn as sns

# Selecciona las columnas pertinentes del dataset
datos = peliculas_disney[['Índice', 'metascore', 'imdb']]

# Calcula la matriz de correlación entre las columnas seleccionadas
matriz_correlacion = datos.corr()

# Crea un mapa de calor con la matriz de correlación entre el índice, el
# metascore y el imdb y lo muestra
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm',
            linewidths=.5)
plt.title('Mapa de calor de correlación entre el índice, el puntaje\
y el imdb')
plt.show()

# Comprobar que esté en formato pep8
pep8(_ih)
```

## gráfico 09

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico en 3D. Coloque título al gráfico y etiquete los ejes.

```
# Importar librerías
from mpl_toolkits import mplot3d

# Selecciona las columnas pertinentes del dataset
index = peliculas_disney['Índice']
puntaje = peliculas_disney['metascore']
internet_movie_data_base = peliculas_disney['imdb']

# Crea una figura y un eje 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotea los datos en 3D
ax.scatter(index, puntaje, internet_movie_data_base, c='r', marker='o',
           label='Datos en 3D')

# Etiqueta los ejes
ax.set_xlabel('Índice')
ax.set_ylabel('Puntaje')
ax.set_zlabel('IMDB')

# Coloca un título al gráfico
plt.title('Gráfico en 3D entre el índice, el puntaje y el imdb')

# Muestra la leyenda
ax.legend()

# Muestra el gráfico
plt.show()

# Comprobar que esté en formato pep8
pep8(_ih)
```
